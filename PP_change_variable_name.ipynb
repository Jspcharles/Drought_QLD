```python
import xarray as xr
import glob
import os

# Folder containing the NetCDF files
folder = r"D:\Project Works\Drought_QLD\Data\SPI\SPI\spi_12"

# Find all .nc files
files = glob.glob(os.path.join(folder, "*.nc"))

for file in files:
    print(f"Processing: {os.path.basename(file)}")

    try:
        # Open and fully load the dataset into memory
        with xr.open_dataset(file) as ds_open:
            ds = ds_open.load()

        print("  Variables found:", list(ds.data_vars))

        # Check whether the old variable exists
        if "spi_scale12" not in ds.data_vars:
            print("  Variable not found")
            continue

        # Rename only the variable
        ds = ds.rename({"spi_scale12": "spi_12"})

        # Overwrite the original file
        ds.to_netcdf(file, mode="w")

        ds.close()

        print("  ✓ Renamed spi_scale12 to spi_12")

    except Exception as error:
        print(f"  ✗ Error: {error}")

print("Done!")
```
